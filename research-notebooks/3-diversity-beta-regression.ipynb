```r
library(betareg)
```

```r
data = read.csv('/home/matt/diversity_data.csv')
```

```r
res <- betareg(diversity_25 ~ crowd_pct + crowd_pct_sq  + project_age + total_issues + total_contributors
               + gini_coefficient + avg_clustering + avg_min_path  + topic_0 + topic_1 + topic_2 
               + topic_3 + topic_4 + topic_5 + topic_6 + topic_7 + topic_8 + topic_9 + topic_10 + topic_11 
               + topic_12 + topic_13 + topic_14 + topic_15 + topic_16 + topic_17 + topic_18 + topic_19 
               + topic_20 + topic_21 + topic_22 + topic_23 , data=data)
```

```r
summary(res)
```

```r
residuals <- resid(res)
```

```r
plot(data$diversity_25, residuals, 
     ylab="Residuals", xlab="Waiting Time", 
     main="Old Faithful Eruptions")
```
